**Importing The Required Libraries**

```python
import easyocr
import PIL
from PIL import ImageDraw
import re
import cv2 
import numpy as np
from matplotlib import pyplot as plt
import shutil
import os
import pandas as pd
import numpy as np
```

```python
# EXECUTE IF INPUT IS PDF
from pdf2image import convert_from_path
 
 
# Store Pdf with convert_from_path function
images = convert_from_path('/content/drive/MyDrive/FS/Callon.pdf')

os.chdir('/content/drive/MyDrive/FS/Images')
for i in range(len(images)):
   
      # Save pages as images in the pdf
    images[i].save('page'+ str(i) +'.jpg', 'JPEG')
```

```python
#Generating outputs for all the images

#create an output folder named results in the datset directory

img_dir="/content/drive/MyDrive/FS/Images"
out_dir="/content/drive/MyDrive/FS/CSV/"
images=os.listdir(img_dir)

df=pd.DataFrame()

reader = easyocr.Reader(['en']) 

for i in range(len(images)):
  # this needs to run only once to load the model into memory
  result = reader.readtext(os.path.join(img_dir,images[i]))
  csv_list=[]
  #print(images[i])
  for j in range(len(result)):

    #extracting text and co-ordinates

    arr=np.array(result[j][0])
    arr=arr.flatten()
    arr=arr.tolist()
    arr.append(result[j][1])
    csv_list.append(arr)
  dfx = pd.DataFrame(csv_list)
  newdfx = dfx.assign(page = i)
  df = pd.concat([df,newdfx])

    
    #making the dataframe
#df=pd.DataFrame(csv_list)
df.columns=df.iloc[0]
df = df[1:]
  #print(df.head())
df.to_csv(out_dir+'FSX.csv',index=False)

```
