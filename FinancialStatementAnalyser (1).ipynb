```python
#doctype: 0-Invoice 1-Financial Statement
doctype = 1
df.columns = ['x1','y1','x2','y2','x3','y3','x4','y4','label','page']
df['label'] = df['label'].str.lower()
#print(df)
#print(df["label"])
labels = pd.DataFrame()
labelsyr = pd.DataFrame()
labelsb = pd.DataFrame()

y_content= ['cash', 'debt', 'income', 'assets','equity']#contains elements to search on y axis (top to bottom) and find data on grid with x_content
x_content = ['2021','2020']#contains elements to search on x axis (left to right) and find data on grid with y_content
beside_content = []#contains elements to search and obtain data that is beside them

for i in y_content:
  
  temp = df.loc[df['label'].str.contains(i)]
  labels = pd.concat([labels,temp])
#print(labels)


for i in x_content:
  
  tempyr = df.loc[df['label']==i] #.str.contains(i)]
  labelsyr = pd.concat([labelsyr,tempyr])
#print(labelsyr)

if doctype == 0:
  for i in beside_content:
    
    tempb = df.loc[df['label'].str.contains(i)]
    labelsb = pd.concat([labelsb,tempb])
  #labelsb = labelsb.loc[~labelsb['label'].str.contains(' ')] #USE ONLY IF THE FIELDS SHOULD BE PERFECTLY EQUAL IE. HAVE NO OTHER WORDS WITH THEM
  print(labelsb)
  labelsb.reset_index()

labels.reset_index()
labelsyr.reset_index()


newl = pd.DataFrame()
newlyr = pd.DataFrame()


for ind in labels.index:
  for ind2 in labelsyr.index:

    x = labels['page'][ind]==labelsyr['page'][ind2]
    
    if not type(labels['page'][ind]==labelsyr['page'][ind2]) is np.bool_ : #IF 2 ENTRIES HAV SAME INDICES, THIS MAY RETURN A SERIES WHICH NEEDS TO BE HANDLED PROPERLY
      x = x.all()
      print(x)
    if(x):
      
      x1 = labels.loc[labels.index == ind]
      newl= pd.concat([newl,x1])

      x2 = labelsyr.loc[labelsyr.index == ind2]
      newlyr= pd.concat([newlyr,x2])
      newl = newl.drop_duplicates(keep = 'first')
      newlyr = newlyr.drop_duplicates(keep = 'first')

      


print(newl)
print(newlyr)

```

```python
def compc(x,y):
  c= 20
  c1 = y + c
  c2 = y - c
  if (c2 < x) and (c1 > x):
    return True
  else:
    return False

```

```python
#EXECUTE IF X AND Y GRID VALUES ARE NEEDED

needed = []

for ind in newl.index:
  for ind2 in newlyr.index:
    
     
    if (newl['page'][ind]==newlyr['page'][ind2]):
      
      y1 = newl['y1'][ind]
      y3 = newl['y3'][ind]
      x1 = newlyr['x1'][ind2]
      x3 = newlyr['x3'][ind2]
      
      valuedf = df.loc[df['page']==newl['page'][ind]]
      valuedf['label'] = valuedf['label'].replace([','],'', regex =True)
      #valuedf = valuedf.loc[valuedf['label'].astype(str).str.isdigit()]
      
      for i in valuedf.index:
        x1v = valuedf['x1'][i]
        x3v = valuedf['x3'][i]
        y1v = valuedf['y1'][i]
        y3v = valuedf['y3'][i]
        ans = False
        
        ans =  compc(y1v, y1) and compc(y3v, y3) and ( compc(x1v, x1) or compc(x3v, x3))
        
        

        
        if ans:
          needed.append([newl['label'][ind], valuedf['label'][i], newlyr['label'][ind2], newl['page'][ind] ])


print(needed)

final = pd.DataFrame(needed)
final.columns = ['label', 'data','year','page']

print(final.loc[final['year']== '2020'])
print('_________________________________________________________________________________________________________')
print(final.loc[final['year']== '2021'])

      
```

```python
if doctype ==1:
  formulalist = [0] * len(y_content)
  for i in range(len(y_content)):
    for j in needed:
      if y_content[i] in j[0]:
        if i < len(formulalist):
          formulalist[i] = float(j[1])
        else:
          if(float(j[1]) > float(formulalist[i])):
            formulalist[i] = float(j[1])

  fs = formulalist
  print(fs)
  if(doctype == 1):
    ebitda = (fs [0] + fs [1] + fs [2] + fs [3] )/ fs[4]
    print('EBITDA = '+ str(ebitda)) #MORE RATIOS CAN BE ADDED
    print('\nCompany Profitablity:')
    if(ebitda< 0.03):
      print('Sick')
    elif(ebitda< 0.05):
      print('Unhealthy')
    elif(ebitda< 0.08):
      print('Average')
    elif(ebitda< 0.11):
      print('Healthy')
    else:
      print('Great')

  elif(doctype == 0):
    for i in needed:
      print(str(i[0]) + '\t' + str(i[1]))
```
